```python
def menu():

    print("Please put your info for your vehicle.")
    print("If it is a Car (enter 1), if Truck (enter 2) if Motorcycle (enter 3) and if you want to input all (enter 0): ")
    dir = input("Enter Here: ")

    match dir:
        case '1':
            cmake = input("Enter Car Make: ")
            cmodel = input("Enter Car Model: ")
            cyear = input("Enter Car Year (Four Digits): ")
            cdoors = input("Enter Car Number of Doors: ")
            chold = input("Enter Car Hold: ")

            Car = car(cmake, cmodel, cyear, chold, cdoors)

            print("\nCar Information:")
            Car.printinfo()
            
        case '2': 
            tmake = input("Enter Truck Make: ")
            tmodel = input("Enter Truck Model: ")
            tyear = input("Enter Truck Year (Four Digits): ")
            tfuel = input("Enter Truck Fuel Type (0 = Gas, 1 = Diesel): ")
            if tfuel != 1 or tfuel != 0:
                raise Exception("Sorry, wrong numbers")
            thold = input("Enter Truck Hold: ")

            Truck = truck(tmake, tmodel, tyear, tfuel, thold)

            print("\nTruck Information:")
            Truck.printinfo()
    
        case '3':
            mmake = input("Enter Motorcycle Make: ")
            mmodel = input("Enter Motorcycle Model: ")
            myear = input("Enter Motorcycle Year (Four Digits): ")
            mcolor = input("Enter Motorcycle Color: ")
            mhold = input("Enter Motorcycle Hold: ")

            Motor = motorcycle(mmake, mmodel, myear, mcolor, mhold)

            print("\nMotorcycle Information:")
            Motor.printinfo()

        case '0':
            cmake = input("Enter Car Make: ")
            cmodel = input("Enter Car Model: ")
            cyear = input("Enter Car Year (Four Digits): ")
            cdoors = input("Enter Car Number of Doors: ")
            chold = input("Enter Car Hold: ")

            tmake = input("Enter Truck Make: ")
            tmodel = input("Enter Truck Model: ")
            tyear = input("Enter Truck Year (Four Digits): ")
            tfuel = input("Enter Truck Fuel Type (0 = Gas, 1 = Diesel): ")
            if tfuel != 1 or tfuel != 0:
                raise Exception("Sorry, wrong numbers")
                    
            thold = input("Enter Truck Hold: ")

            mmake = input("Enter Motorcycle Make: ")
            mmodel = input("Enter Motorcycle Model: ")
            myear = input("Enter Motorcycle Year (Four Digits): ")
            mcolor = input("Enter Motorcycle Color: ")
            mhold = input("Enter Motorcycle Hold: ")

            Car = car(cmake, cmodel, cyear, chold, cdoors)
            Truck = truck(tmake, tmodel, tyear, tfuel, thold)
            Motor = motorcycle(mmake, mmodel, myear, mcolor, mhold)

            print("\nCar Information:")
            Car.printinfo()

            print("\nTruck Information:")
            Truck.printinfo()

            print("\nMotorcycle Information:")
            Motor.printinfo()
            

        case default:
            print("Invalid")
            menu()
            

class Vehicle:
    
    def __init__(self, vyear, vmake, vmodel):

        self.year = vyear
        self.make = vmake
        self.model = vmodel

    
    def printinfo(self):
        
        m = "Here is the info on your vehicles. \nYear: " + str(self.year)
        s = "\nMake: " + str(self.make)
        g = "\nModel: " + str(self.model)


        return print(m + s + g)
        

class truck(Vehicle):


    def __init__(self, vmake, vmodel, vyear, tfuel, thold):
        super().__init__(vmake, vmodel, vyear)

        self.hold = thold
        
        if tfuel == "0":
            tfuel = "Gas"
        elif tfuel == "1":
            tfuel = "Diesel"
      
            
        self.fuel = tfuel


    def printinfo(self):
        super().printinfo()

        add = "Fuel: " + str(self.fuel) + "\nHold: " + str(self.hold)
        
        return print(add)



class car(Vehicle):

    def __init__(self, vmake, vmodel, vyear, chold, cdoor):
        super().__init__(vmake, vmodel, vyear)

        self.hold = chold
        self.door = cdoor

    def printinfo(self):
        super().printinfo()
        
        add = "Holds: " + str(self.hold) + "\nDoors: " + str(self.door)
        return print(add)



class motorcycle(Vehicle):

    def __init__(self, vmake, vmodel, vyear, mcolor, mhold):
        super().__init__(vmake, vmodel, vyear)
        
        self.color = mcolor
        self.hold = mhold

    def printinfo(self):
        super().printinfo()
        
        add = "Color: " + str(self.color) + "\nHolds: " + str(self.hold)
        return print(add)


def main():

    menu()


if __name__ == "__main__":
    main()
```
